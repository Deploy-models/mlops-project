## Q5. Training a model
Now let's use the feature matrix from the previous step to train a model.

```python
#Split into train, val, test set
from sklearn.model_selection import train_test_split
target = "duration"
y_target = df_clean["duration"]

# 1. First split: Separate out the Test set (10% of total)
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_train, y_target, test_size=0.1, random_state=42
)

# 2. Second split: Split the remaining "train_full" into Train and Val
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.11, random_state=42
)

print(f"Train: {X_train.shape[0]}, Val: {X_val.shape[0]}, Test: {X_test.shape[0]}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
rmse = root_mean_squared_error(y_test, y_pred)
print(f"The RMSE on test is: {rmse:.2f}")
```

```python
df_val = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-02.parquet")
df_val.head()
```

```python
df_val.to_csv("yellow_tripdata_2024-02.csv")
```

```python
def read_dataframe(filename):
    df = pd.read_csv(filename)

    # Q2: Computing duration in minutes
    df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
    df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
    
    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.dt.total_seconds() / 60

    # Q3: Dropping outliers (inclusive)
    df = df[(df.duration >= 1) & (df.duration <= 60)]

    # Q4 preparation: Casting IDs to strings
    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    
    return df
```

```python
# 1. Load and clean the data
df_train = read_dataframe('yellow_tripdata_2024-01.csv')
df_val = read_dataframe('yellow_tripdata_2024-02.csv')

# 2. Extract features as a list of dictionaries
categorical = ['PULocationID', 'DOLocationID']
train_dicts = df_train[categorical].to_dict(orient='records')
val_dicts = df_val[categorical].to_dict(orient='records')

# 3. Q4: One-hot encoding using DictVectorizer
dv = DictVectorizer()

# Fit and transform on training data
X_train = dv.fit_transform(train_dicts)

# ONLY transform on validation data (to keep the same columns)
X_val = dv.transform(val_dicts)

# 4. Get the targets
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values

print(f"Feature matrix dimensionality: {X_val.shape[1]}")
```

```python
lr_new = LinearRegression()
lr_new.fit(X_train, y_train)
```

```python
y_pred = lr_new.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
print(f"The RMSE on test is: {rmse:.2f}")
```
